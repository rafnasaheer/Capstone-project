```python
df.drop('duration',axis=1)
```

```python
x=df.drop(['y'],axis=1)
y=df['y']
```

```python
x=pd.get_dummies(x,drop_first='true',dtype=int)
```

```python
#logistic regression
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(multi_class='ovr',max_iter=1000)
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred,output_dict=True)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
#ensemble modelling
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
SM=SMOTE(random_state=42)
x_resample,y_resample=SM.fit_resample(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
cls=RandomForestClassifier()
```

```python
n_estimators=[25,100]
criterion=['gini','entropy']
max_depth=[5,10,25]
```

```python
parameters={'n_estimators':n_estimators,'criterion':criterion,'max_depth':max_depth}
```

```python
RF_CLS=GridSearchCV(cls,parameters)
RF_CLS.fit(x_resample,y_resample)
```

```python
y_pred_RF=RF_CLS.predict(x_test)
y_pred_RF
accuracy_score(y_test,y_pred_RF)*100
```

```python
RF_CLS.best_params_
```

```python
cls=RandomForestClassifier(criterion='gini', max_depth=25, n_estimators=100)
cls.fit(x_resample,y_resample)
```

```python
y_pred_CLS=cls.predict(x_test)
y_pred_CLS
```

```python
accuracy_score(y_test,y_pred_CLS)*100
```

```python
confusion_matrix(y_test,y_pred_CLS)
```

```python
classification_report(y_test,y_pred_CLS,output_dict=True)
```
