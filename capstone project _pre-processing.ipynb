```python
job_success = df.groupby('job')['y'].value_counts(normalize=True).unstack()
job_success['success_rate'] = job_success['yes'] / (job_success['yes'] + job_success['no'])
job_success = job_success.sort_values('success_rate', ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=job_success.index, y='success_rate', data=job_success)
plt.xticks(rotation=45)
plt.title('Subscription Rate by Job')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='y', y='duration', data=df)
plt.title('Call Duration Distribution by Subscription')
plt.show()
```
